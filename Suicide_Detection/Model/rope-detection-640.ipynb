```python
from ultralytics import YOLO
from pathlib import Path
```

```python
import os
os.makedirs('/kaggle/working/Model', exist_ok=True)
os.makedirs('/kaggle/working/Model/Model_Data/logs', exist_ok=True)
```

```python
import yaml

model_data_path = '/kaggle/working/Model'
dataset_path = '/kaggle/input/rope-dataset/Processed'

config = {
    'path': dataset_path,
    'train': 'train/images',
    'val': 'val/images',
    'test': 'test/images',
    'names': {0: 'rope'}
}

with open(model_data_path + '/config.yaml', 'w') as f:
    yaml.dump(config, f, default_flow_style=False)
```

```python
model = YOLO('yolov8n.yaml')

model.train(data=model_data_path + '/config.yaml', imgsz=640, batch=64, epochs=50, name='SuicideDetection', save_dir='/kaggle/working/Model/Model_Data/logs')
```
